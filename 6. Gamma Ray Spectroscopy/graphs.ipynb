```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

data = np.loadtxt('endsem.txt')


x = data[:, 0]
y = data[:, 1]

xnew = np.linspace(x.min(), x.max(), 300)

gfg = make_interp_spline(x, y, k=2)

y_new = gfg(xnew)
plt.scatter(x, y, label = "Observed data points")  
plt.plot(xnew, y_new, color = "orange", label = "Trend line")


plt.ylabel("Resolution, (percent)")
plt.xlabel("Operating voltage (V)")

plt.legend()

plt.savefig("plot-endsem.png", dpi = 1200)
plt.show()
```
